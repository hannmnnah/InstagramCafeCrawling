```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import re 
from urllib.request import urlopen
import pandas as pd
import numpy as np
```

```python
def insta_searching(word):  #wordë¼ëŠ” ë§¤ê°œë³€ìˆ˜ë¥¼ ë°›ëŠ” insta_searching ì´ë¼ëŠ” í•¨ìˆ˜ ìƒì„±
    url = 'https://www.instagram.com/explore/tags/' + word
    return url
```

```python
import time 
def select_first(driver):
    first = driver.find_element_by_css_selector('div._9AhH0') 
    #find_element_by_css_selector í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•´ ìš”ì†Œ ì°¾ê¸°
    first.click()
    time.sleep(3)
```

```python
import re 
def get_content(driver,name):
    # 1. í˜„ì¬ í˜ì´ì§€ì˜ HTML ì •ë³´ ê°€ì ¸ì˜¤ê¸°
    time.sleep(np.random.randint(0.3,1))
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')

    
    name = name
    # 2. ë³¸ë¬¸ ë‚´ìš© ê°€ì ¸ì˜¤ê¸°
    try:  #ì—¬ëŸ¬ íƒœê·¸ì¤‘ ì²«ë²ˆì§¸([0]) íƒœê·¸ë¥¼ ì„ íƒ  
        content = soup.select('div.C4VMK > span')[0].text 
        #ì²« ê²Œì‹œê¸€ ë³¸ë¬¸ ë‚´ìš©ì´ <div class="C4VMK"> ì„ì„ ì•Œ ìˆ˜ ìˆë‹¤.
                                #íƒœê·¸ëª…ì´ div, classëª…ì´ C4VMKì¸ íƒœê·¸ ì•„ë˜ì— ìˆëŠ” span íƒœê·¸ë¥¼ ëª¨ë‘ ì„ íƒ.
    except:
        content = ' ' 
    # 3. ë³¸ë¬¸ ë‚´ìš©ì—ì„œ í•´ì‹œíƒœê·¸ ê°€ì ¸ì˜¤ê¸°(ì •ê·œí‘œí˜„ì‹ í™œìš©)
    tags = re.findall(r'#[^\s#,\\]+', content) # content ë³€ìˆ˜ì˜ ë³¸ë¬¸ ë‚´ìš© ì¤‘ #ìœ¼ë¡œ ì‹œì‘í•˜ë©°, #ë’¤ì— ì—°ì†ëœ ë¬¸ì(ê³µë°±ì´ë‚˜ #, \ ê¸°í˜¸ê°€ ì•„ë‹Œ ê²½ìš°)ë¥¼ ëª¨ë‘ ì°¾ì•„ tags ë³€ìˆ˜ì— ì €ì¥
    # 4. ì‘ì„± ì¼ì ê°€ì ¸ì˜¤ê¸°
    time.sleep(0.4)
    try:
        date = soup.select('time._1o9PC.Nzb55')[0]['datetime'][:10] #ì•ì—ì„œë¶€í„° 10ìë¦¬ ê¸€ì
    except:
        date = ''
    # 5. ì¢‹ì•„ìš” ìˆ˜ ê°€ì ¸ì˜¤ê¸°
    try:
        like = soup.select('div.Nm9Fw > button')[0].text[4:-1] 
    except:
        like = 0
    # 6. ìœ„ì¹˜ ì •ë³´ ê°€ì ¸ì˜¤ê¸°
    try:
        place = soup.select('div.JF9hh')[0].text
    except:
        place = ''
    time.sleep(0.3)
    try: 
        link = soup.select_one('a.sqdOP.yWX7d').get('href')
    except:
        time.sleep(3)
        link =''
    
    # 7. ìˆ˜ì§‘í•œ ì •ë³´ ì €ì¥í•˜ê¸°
    data = {'name' : name,
            'content':content,
            'date' :date,
            'like' : like,
            'place' : place, 
            'tags' : tags,
            'link' : 'https://www.instagram.com' +link}
    
    return data 
  
```

```python
def move_next(driver):
    right = driver.find_element_by_css_selector('a._65Bje.coreSpriteRightPaginationArrow') 
    right.click()

    time.sleep(3)
```

```python
def get_result(name, mail, pw, number):
    
    
    driver = webdriver.Chrome('C:\\Code\\5.1 [í¬ë¡¤ë§ í”„ë¡œì íŠ¸]\\chromedriver')
    url_log = 'https://www.instagram.com/accounts/login/'
    #2. ë¡œê·¸ì¸ í•˜ê¸°
    driver.get(url_log)
    time.sleep(2)
    driver.find_element_by_name("username").send_keys(mail)
    driver.find_element_by_name("password").send_keys(pw)
    driver.find_element_by_css_selector('.sqdOP.L3NKy.y3zKF').click()
    time.sleep(3)
    driver.find_element_by_css_selector('.sqdOP.yWX7d.y3zKF').click()

    time.sleep(4) 

    #3. ê²€ìƒ‰í˜ì´ì§€ ì ‘ì†í•˜ê¸°
    
    url = insta_searching(name)
    driver.get(url)
    time.sleep(4) 

    #4. ì²«ë²ˆì§¸ ê²Œì‹œê¸€ ì—´ê¸°
    select_first(driver) 

    #5. ë¹„ì–´ìˆëŠ” ë³€ìˆ˜(results) ë§Œë“¤ê¸°
    global results
    results = [] 

    #ì—¬ëŸ¬ ê²Œì‹œë¬¼ í¬ë¡¤ë§í•˜ê¸°
    
    for i in range(number):
        data = get_content(driver,name) #ê²Œì‹œë¬¼ ì •ë³´ ê°€ì ¸ì˜¤ê¸°
 
        results.append(data)
        print(i,'/',number)
        time.sleep(np.random.randint(1,3))
        move_next(driver)
        time.sleep(1)
        
    driver.quit()
    
    df_total = pd.DataFrame(results)
    return df_total
```
